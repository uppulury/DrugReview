```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

DR = pd.read_csv('DrugReview.csv')

#Extract Information Regarding the (Major,Moderate,Minor) SideEffects
Major = [0]*len(DR)
Moderate = [0]*len(DR)
Minor = [0]*len(DR)
Interactions = DR['Drug_Interactions']
cnt = 0
for i in range(0,len(Interactions)):
    if type(Interactions[i]) == str:
        Interaction_List = Interactions[i].split(',')
        List_Length = len(Interaction_List)
        c1 = 0
        c2 = 0
        c3 = 0
        for j in range(0,List_Length):
            List_j = Interaction_List[j].split()
            if List_j[1]=='major':
                Major[i] = List_j[0]
                c1 = c1+1
            if List_j[1]=='moderate':
                Moderate[i] = List_j[0]
                c2 = c2+1
            if List_j[1]=='minor':
                Minor[i] = List_j[0]
                c3 = c3+1
        if c1 == 0:
            Major[i] = 'NA'
        if c2 == 0:
            Moderate[i] = 'NA'
        if c3 == 0:
            Minor[i] = 'NA'            
    else:
        Major[i] = 'NA'
        Moderate[i] = 'NA'
        Minor[i] = 'NA'
        
```

```python
#Assign the Major,Moderate,Minor Interaction Fields
DR['Major'] = Major
DR['Moderate'] = Moderate
DR['Minor'] = Minor
```

```python
#Extract and Assign the 'Duration' categorical variable to a Number
#For Example the 'Duration' Field with value NaN gets a '0'.
values = DR['Duration'].unique()
DurationNew = [0]*len(DR)
for i in range(1,len(DR['Duration'].unique())):
    indices = list(DR.index[DR['Duration']==values[i]])
    for index in indices:
        DurationNew[index] = i
        
#Categorize 'DurationNew' in the dataframe 'DR'
DR['DurationCategory'] = DurationNew
```
